### 3.2 Exploratory Data Analysis Practice Exercise - 1.

```python
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
df['fare_per_year'] = df['Fare'] / df['Age'] #add column fare per year
high_fare_age = df[df['fare_per_year']>5] #subest rows where fare per year is greater than 5
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year' ,ascending=False) # sort values in descending
result = high_fare_age_srt[['Name', 'fare_per_year']] #select columns name and fare per year
result
```

### Which passenger had the highest fare paid relative to their age?

```python
highest_fare_per_class = high_fare_age_srt.iloc[0]
highest_fare_per_class
```

```python
df['fare_per_class'] = df['Fare']/ df['Pclass']
adult_males = df[(df['Sex']=='male') & (df['Age']>=18)]
adult_males_srt = adult_males.sort_values(by='fare_per_class',ascending=False)
result = adult_males_srt[['Name','Age','fare_per_class']]
result
```

### Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their class?

```python
highest_adult_fare_per_class = adult_males_srt.iloc[0]
highest_adult_fare_per_class
```

### 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

```python
total_fare = df['Fare'].sum()
first_class_passenger = df[df['Pclass']==1]
first_class_passenger_fare =first_class_passenger['Fare'].sum()
second_class_passenger = df[df['Pclass']==2]
second_class_passenger_fare =second_class_passenger['Fare'].sum()
third_class_passenger = df[df['Pclass']==3]
third_class_passenger_fare =third_class_passenger['Fare'].sum()
fare_totals = [first_class_passenger_fare,second_class_passenger_fare,third_class_passenger_fare]/total_fare
fare_totals
```

### What percent of the total fare revenue came from each passenger class?

```python
fare_percentage = [float(p) * 100 for p in fare_totals]
fare_percentage
```

```python
df['age_group'] = pd.cut(df['Age'], bins=[0, 18, 64, float('inf')],labels=['child', 'adult', 'senior'],right=False)
total_passengers = len(df)
age_group_counts = df['age_group'].value_counts()

proportion=age_group_counts/total_passengers
proportion_percentage = proportion * 100
proportion_percentage
```

### What percent of the total number of passengers on the Titanic belonged to each age group(e.g., child, adult, senior)?

```python
proportion_percentage #total number belonged to each age group
```
